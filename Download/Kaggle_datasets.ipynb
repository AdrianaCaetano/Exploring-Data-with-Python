## Download Kaggle datasets

To download a competition dataset: 
    `! kaggle competitions download <name-of-competition>`

To download specific datasets:
    `! kaggle datasets download <www.kaggle.com/user-name/dataset-name>`

To download a specific file: 
    `! kaggle competitions/datasets download <name-of-competition/name-of-dataset> -f <file-name.csv>`

In case you get a zip file:
    `! unzip <name-of-file>`
